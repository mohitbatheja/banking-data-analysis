# Average Call Duration by Month #

```python
plt.figure(figsize=(10,6))

month_duration = df.groupby('month')['duration'].mean().reset_index()

sns.lineplot(data=month_duration, 
             x='month', 
             y='duration', 
             marker='o', 
             markersize = 10,
             markerfacecolor = 'red',
             markeredgecolor = 'yellow',
             label = "Average Duration")
ax = plt.gca()

for spines in ax.spines.values():
    spines.set_color("black")
    
ax.spines['top'].set_color('white')
ax.spines['right'].set_color('white')

ax.tick_params(axis = 'both', colors ='Black',length=7)

for x_val, y_val in zip(month_duration["month"], month_duration["duration"]):
    plt.text(
        x_val,
        y_val,
        round(y_val, 1),   # 1 decimal place
        ha="center",
        va="bottom",
        fontsize=10,
        fontweight="bold",
        color="black"
    )

plt.title("Average Call Duration by Month",fontweight = 'bold', fontsize = 15, color = 'black')
plt.legend(shadow = True,facecolor = 'lightgrey',edgecolor = 'black')
plt.xlabel("Month")
plt.ylabel("Average Duration")
plt.grid(alpha = 0.1)
plt.show()
```

**Insight**<br>
- **Maximum Peak:** December records the highest average call duration at **345.6**.
- **Minimum Trough:** August sees the lowest average call duration drop to **232.5**.
- **Mid-Year Plateau:** Call lengths mostly stabilize between June and November, hovering around **242 to 261**.
- **Late-Sequence Rise:** A steady upward trend occurs from November (**245.9**) through October (**288.6**) to September (**297.7**).
- **Non-Chronological Axis:** The X-axis reflects a custom or sorted categorical ordering rather than a standard calendar timeline.

```python
plt.figure(figsize=(8,6))

sns.regplot(x='age', y='duration', data=df)

plt.title("Regression: Age vs Call Duration")
plt.xlabel("Age")
plt.ylabel("Call Duration")
plt.show()
```

**Insight**<br>
- **Weak or No Linear Relationship:** The regression line is virtually flat, indicating that an individual's age does not linearly predict or influence their call duration.
- **Concentrated Data Density:** The vast majority of calls are heavily concentrated among individuals aged 25 to 60, with call durations mostly staying below 1,000 units.
- **Presence of Outliers:** High-duration call outliers (reaching up to 3,000–5,000 units) primarily occur within the 30 to 55 age bracket.
- **Data Sparsity in Older Demographics:** As age increases beyond 60, the volume of data points thins out significantly, though call durations consistently remain low.
